Pour afficher ce cluestering je voulais initialement afficher les résultats dans un dendrogramme, mais pour des soucis de visibilité j'ai opté pour une carte.

Code pour le deuxième graphique :

```python


#je charge les données
df = pd.read_csv('mismanaged_plasticwaste.csv')

#Dans mon fichier csv je n'ai pas de code Pays, j'ai donc ajouté cette fonction pour rajouter ce code.
def findCountry(country_code):
    try:
        return pycountry.countries.get(name=country_code).alpha_3
    except:
        return 'NaN'

df['country_code'] = df.apply(lambda row: findCountry(row.Country), axis=1) #Je rajoute les codes pays.


#Cluestering

#J'enlève les valeurs non numérique de mon dataframe
df1 = df.drop(['Country', 'country_code'], axis=1)

#
kmeans_model = KMeans(n_clusters=5, random_state=10).fit(df1)
labels = kmeans_model.labels_
df1['cluster'] = labels #Ici je créer un 2nd dataframe avec les countrycode pour pouvoir afficher ma carte correctement.
df1['Country'] = df['Country']
df1['country_code'] = df['country_code']

# Créer un GeoDataFrame sans coordonnées géographiques
gdf = gpd.GeoDataFrame(df1)

# Charger une carte du monde avec GeoPandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Fusionner votre DataFrame avec la carte du monde
merged = world.set_index('iso_a3').join(gdf.set_index('country_code'))

# Tracer la carte
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1, color='black')
merged.plot(column='cluster', cmap='Set1', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Map by Clusters')

# Afficher la carte
plt.show()
```


Lors de mes études antérieures, j'ai identifié une corrélation entre le taux de plastique gaspillé ou mal utilisé et le PIB des pays. Cependant, j'ai observé une agrégation inappropriée des pays d'Afrique et d'Europe dans le même cluster, malgré des différences significatives dans leurs infrastructures et leur sensibilisation à la pollution. Ce regroupement peut être attribué à un nombre de clusters trop bas (5). En augmentant ce nombre à 7, j'ai réussi à séparer plus clairement les pays africains des pays européens, mais cela a eu pour conséquence de perdre des informations sur les pays asiatiques.

Par exemple, les pays d'Asie du Sud se retrouvent regroupés, ce qui est cohérent avec leur notoriété en termes de rejets de plastique dans les cours d'eau. Cette constatation souligne le défi délicat de choisir le bon nombre de clusters, car un nombre élevé peut conduire à une meilleure séparation entre certains groupes, mais au détriment de la clarté et de la pertinence globale de l'analyse.

Afin de trouver le juste équilibre, des ajustements continus du nombre de clusters pourraient être nécessaires, en tenant compte des caractéristiques spécifiques des différentes régions géographiques. Il est crucial de reconnaître que la complexité des facteurs influençant la gestion des déchets plastiques nécessite une approche nuancée pour interpréter correctement les résultats des analyses de clustering.
